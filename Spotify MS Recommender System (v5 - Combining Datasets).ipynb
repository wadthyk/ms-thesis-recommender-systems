# Standardizing the 2 datasets before unifying them

```python
# function to process all changes at once

import pandas as pd

def process_dataframes(*dataframes, new_order, drop_column):
    """
    Process multiple DataFrames by changing column order and dropping a specified column.
    
    Parameters:
    *dataframes: Variable number of pandas DataFrames to process
    new_order: List of column names in the desired order
    drop_column: Name of the column to drop
    
    Returns:
    List of processed DataFrames
    """
    processed_dfs = []

    for df in dataframes:
        # Drop the specified column
        df = df.drop(columns=[drop_column])

        # Reorder the columns
        df = df[new_order]

        processed_dfs.append(df)

    return processed_dfs

# Define the new column order (excluding the column to be dropped)
new_column_order = ['Artist', 'Title', 'URI', 'track_id', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'type', 'id', 'uri', 'track_href', 'analysis_url', 'duration_ms', 'time_signature', 'MBTI']

# Specify the column to drop
column_to_drop = 'Album'
```

```python
# assigning temporary new df_names

df1 = df_wadthy_liked
df2 = df_withy_liked
df3 = df_yeonju_liked
df4 = df_simon_liked
df5 = df_soeren_liked
df6 = df_nga_liked
df7 = df_han_liked
df8 = df_yoojin_liked
```

```python
# Process all DataFrames at once
processed_df1, processed_df2, processed_df3, processed_df4, processed_df5, processed_df6, processed_df7, processed_df8 = process_dataframes(df1, df2, df3, df4, df5, df6, df7, df8,
                                                                                                                                            new_order=new_column_order,
                                                                                                                                            drop_column=column_to_drop)
```
