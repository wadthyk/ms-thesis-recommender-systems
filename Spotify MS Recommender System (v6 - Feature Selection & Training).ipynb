### Random Forest

```python
# Select features and target
features = df_withy_unified[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature']]
target = df_withy_unified['like_dislike']  # binary column indicating like (1) or dislike (0)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Initialize the model
from sklearn.ensemble import RandomForestClassifier

# random forest classifier
model = RandomForestClassifier(random_state=42)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Make predictions
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Classification Report:\n{class_report}')
```

### K-Nearest Neighbor (KNN)

```python
# Select features and target
features_knn = df_withy_unified[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature']]
target_knn = df_withy_unified['like_dislike']  # binary column indicating like (1) or dislike (0)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features_knn, target_knn, test_size=0.2, random_state=42)
```

```python
# KNN is sensitive to the scale of the data, so it’s important to standardize the features.
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the training data
X_train_scaled = scaler.fit_transform(X_train)

# Transform the testing data
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize the model
knn = KNeighborsClassifier(n_neighbors=5)  # can be tuned to the 'n_neighbors' parameter
```

```python
# Train the model
knn.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Make predictions
y_pred = knn.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Classification Report:\n{class_report}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function to create radar chart
def create_radar_chart(ax, data, features, title):
    # Number of variables we're plotting
    num_vars = len(features)

    # Compute angle of each axis
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

    # The plot is a circle, so we need to "complete the loop"
    angles += angles[:1]

    # Draw one axe per variable and add labels
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(features)

    # Draw y-labels
    ax.set_rscale('linear')
    ax.set_rlabel_position(0)
    ax.set_yticks([0.2, 0.4, 0.6, 0.8])
    ax.set_yticklabels(["0.2", "0.4", "0.6", "0.8"])
    ax.set_ylim(0, 1)

    # Plot data
    values = data[features].values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=1, linestyle='solid')
    ax.fill(angles, values, 'b', alpha=0.1)

    # Add title
    ax.set_title(title, size=10, color='b', y=1.1)

# Create a grid of radar charts
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20, 20), subplot_kw=dict(polar=True))
fig.tight_layout(pad=5.0)

# Plot each MBTI type
for i, (mbti_type, ax) in enumerate(zip(mean_values['mbti'], axes.flatten())):
    create_radar_chart(ax, mean_values[mean_values['mbti'] == mbti_type], audio_features, mbti_type)

plt.show()
```

### Scatter Plot Distribution of Function Pairs Based on Song Mood

```python
# Define mood categories based on valence and energy
def categorize_mood(row):
    if row['valence_mean'] >= 0.5 and row['energy_mean'] >= 0.5:
        return 'Happy/Joyful'
    elif row['valence_mean'] < 0.5 and row['energy_mean'] >= 0.5:
        return 'Angry/Turbulent'
    elif row['valence_mean'] < 0.5 and row['energy_mean'] < 0.5:
        return 'Sad/Depressing'
    else:
        return 'Chill/Calming'

df_mbti_unified['mood'] = df_mbti_unified.apply(categorize_mood, axis=1)
```

```python
# Count the number of function pairs in each mood category
mood_counts = df_mbti_unified.groupby(['function_pair', 'mood']).size().reset_index(name='count')

print(mood_counts)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the scatter plot
plt.figure(figsize=(10, 10))
sns.scatterplot(data=df_mbti_unified, x='valence_mean', y='energy_mean', hue='function_pair', style='mood', palette='Set2')

# Add mood category labels
plt.axhline(0.5, color='black', linewidth=1)
plt.axvline(0.5, color='black', linewidth=1)
plt.text(0.25, 0.75, 'Angry/Turbulent', horizontalalignment='center', verticalalignment='center', fontsize=12)
plt.text(0.75, 0.75, 'Happy/Joyful', horizontalalignment='center', verticalalignment='center', fontsize=12)
plt.text(0.25, 0.25, 'Sad/Depressing', horizontalalignment='center', verticalalignment='center', fontsize=12)
plt.text(0.75, 0.25, 'Chill/Calming', horizontalalignment='center', verticalalignment='center', fontsize=12)

# Add legend
plt.legend(title='Function Pair', loc='upper right')

# Add title and labels
plt.title('Scatter Plot of Song Mood by Valence and Energy')
plt.xlabel('Valence')
plt.ylabel('Energy')

plt.show()
```
